```python
import numpy as np
import pandas as pd
```

```python
def fetch_financial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('META')
df.head()
```

```python
df.to_csv('fb.csv')
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2023)]
df_nov
```

```python
df_nov.to_csv('fb_nov.csv')
```

```python
new_df = pd.read_csv('fb_nov.csv', index_col=0)
new_df
```

```python
df_nov.to_excel('fb_nov.xlsx')
```

```python
new_df = pd.read_excel('fb_nov.xlsx', index_col=0)
new_df
```
