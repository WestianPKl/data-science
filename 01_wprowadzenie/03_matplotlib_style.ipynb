```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)

_ = plt.plot(x, x_sin)
```

```python
plt.style.use('ggplot')
_ = plt.plot(x, x_sin)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-ticks')
_ = plt.plot(x, x_sin)
```

```python
plt.style.use('dark_background')
_ = plt.plot(x, x_sin)
```

```python
plt.style.use('seaborn-v0_8')
_ = plt.plot(x, x_sin)
```
